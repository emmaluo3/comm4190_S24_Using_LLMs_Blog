As a university student from the middle of nowhere in New Jersey, I unfortunately did not get an invite to Fashion Week this year... however, I decided to create my own runway show based on the most recent 2024 fashion trends! I (again unfortunately) do not have the budget to hire models, find a runway venue, and hire designers, so I decided to create my runway show using AI.
I've taken the hottest trends indicated by https://www.glamour.com/story/2024-fashion-trends and used the description of the trend as the prompt, followed by an indication to generate an image that creates a design based on the trend for a luxury fashion model. In this blog, I'm going to be judging each image based on 4 criteria:
1. Originality - is this something new that we haven't seen before or is this simply a repeat of what has already been done?
2. Accuracy - is the prompt fully understood and displayed?
3. Image quality - is overall image quality - is this something that can be displayed in a commercial or advertisement?
4. Design - the simplest question: is this hot or not?
   

# Bows
## Overall: 8/10

<img src="fashion1.png" style="width:50%;" />

### Originality: 8/10
This generates a unique cultural interpretation of the bow trend. The traditional bow trend is coquette and dainty, however this design creates a bold and strong impression. However, I would've like to see more implementation of the bow trend instead of just "slapping a bow on it"
### Accuracy: 8/10
This implementation of this trend is pretty simple -- just add bows! However, I feel like this design is doing the bare minimum. Bows could be implemented more throughout the designs and be more apparent that they are highlighting this trend.
### Image Quality: 8/10
I like how the image highlights potential pieces that could be used in this design. However, this image seems more artistic/abstract rather than realistic, so the usefulness of this image depends on what you want to do with it.
### Design: 8/10
The design is intricate and does a good job with incorporating pleats and a unique sleeve shape. All the parts are well coordinated and the accessories are paired well. From a first glance, you can't exactly tell what trend the design is trying to encapsulate but it is still a well-designed piece overall.


# Rosettes
## Overall: 7.5/10

<img src="fashion2.png" style="width:50%;" />

### Originality: 6/10
The placement of the rosettes on the blouse is unique. It the most flattering... well we'll talk about that below. However, the blouse itself is quite plain and similar to what we've seen before.
### Accuracy: 10/10
With the abundance of rosettes on the model, the prompt was *definitely* understood.
### Image Quality: 8/10
The model highly resembles a human. The clothing and venue could be improved to be more realistic rather than artistic. Also, it is kind of creepy that all audience members are facing the same way...
### Design: 6/10
I'm going to be straight up here. It seems as if someone glued on roses to a blouse and called it a design. The skirt is slightly better but doesn't seem creative whatsoever.

# Sequins
## Overall: 8.75/10

<img src="fashion3.png" style="width:50%;" />

### Originality: 8/10
Sequins have been around for many years, and I would've liked to see them be used in a way that isn't just "let's put sequins on something to make it sparkly". However, I personally *really* like the bag that was generated, so bonus points for that.
### Accuracy: 10/10
It's pretty clear that sequins were used in this design!
### Image Quality 9/10

### Design: 8/10
I really like the shape of the sleeves and the cut of the dress. The added sheerness at the bottom and sides also gives it some variation. The only thing I don't like are the boots, which seem a bit dated.
